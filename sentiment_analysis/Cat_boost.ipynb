```python
df= pd.read_csv('preprocessed_data_1.csv')
df = df.applymap(lambda x: " ".join(x.split()) if isinstance(x, str) else x)
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
from catboost import CatBoostClassifier, Pool
```

```python
dataset = pd.read_csv("preprocessed_data_1.csv")
dataset = dataset[dataset['sentiment_score'] != 0] 
train, test = train_test_split(dataset, test_size=0.2)
```

```python
df_train = df.loc[train.index][['message_text', 'sentiment_score']]
df_test = df.loc[test.index][['message_text', 'sentiment_score']]
```

```python
y_train, X_train = df_train['sentiment_score'], df_train.drop(['sentiment_score'], axis=1)
```

```python
y_test, X_test = df_test['sentiment_score'], df_test.drop(['sentiment_score'],axis=1)
```

```python
train_pool = Pool(
    data=X_train,
    label=y_train,
    text_features=['message_text']
)
test_pool = Pool(
    data=df_test[[c for c in df_test.columns if c!='sentiment_score']][['message_text']], 
    label=df_test['sentiment_score'],
    text_features=['message_text']
)
```

```python
torch.cuda.is_available
```

```python
from sklearn.metrics import accuracy_score
```

```python
from hyperopt import hp, fmin, tpe, Trials, STATUS_OK
from hyperopt.pyll.base import scope

def objective(params):
    model = CatBoostClassifier(**params, task_type='GPU')
    
    model.fit(train_pool, eval_set=(test_pool), verbose=100)
    
    y_pred = model.predict(test_pool)
    accuracy = accuracy_score(y_test, y_pred)
    
    return {'loss': -accuracy, 'status': STATUS_OK}

# –ü—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–æ –ø–æ–∏—Å–∫–∞ –≤—Å–µ—Ö –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤
space = {
    'depth': hp.quniform('depth', 1, 6, 1),
    'learning_rate': hp.loguniform('learning_rate', -3, 0),
    'l2_leaf_reg': hp.loguniform('l2_leaf_reg', -3, 3),
    'iterations': hp.quniform('iterations', 100, 5000, 100),
    'border_count': hp.quniform('border_count', 1, 255, 1),
    'thread_count': hp.quniform('thread_count', 1, 16, 1),
}


# –ó–∞–ø—É—Å–∫ –æ–ø—Ç–∏–º–∏–∑–∞—Ü–∏–∏ –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤
trials = Trials()
best = fmin(objective, space, algo=tpe.suggest, max_evals=20, trials=trials)

print(best)
```

```python
def fit_model(train_pool, test_pool, **kwargs):
    model_cb = CatBoostClassifier(
        task_type='GPU',
        eval_metric='Accuracy',
        od_type='Iter',
        od_wait=2000,
        border_count =  156.0, 
        depth = 4.0, 
        iterations = 3900.0, 
        l2_leaf_reg = 4.497463825101135, 
        learning_rate = 0.18921853606460642, 
        thread_count = 4.0,
        **kwargs
    )
    return model_cb.fit(
        train_pool,
        eval_set=test_pool,
        verbose=100,
        plot=True,
        use_best_model=True)
```

```python
model_cb = fit_model(train_pool,test_pool)
```

```python
model = fit_model(
    train_pool,test_pool,
    learning_rate=0.35,
    tokenizers=[
        {
            'tokenizer_id': 'Space',
	        'delimiter': ' ',
	        'separator_type': 'ByDelimiter',
            'lowercasing': 'True',
            'token_types':['Word', 'Number', 'SentenceBreak'],
            'sub_tokens_policy':'SeveralTokens'
            

            # 'tokenizer_id': 'Sense',
            # 'separator_type': 'BySense',
            # 'lowercasing': 'True',
            # 'token_types':['Word', 'Number', 'SentenceBreak'],
            # 'sub_tokens_policy':'SeveralTokens'
        }      
    ],
################
    # dictionaries= [
    #     {
    #         'dictionary_id': 'Word',
    #         'Word:min_token_occurrence': '5',
    #         'BiGram:gram_order': '2'
    #     }
    #     ],
##########
    dictionaries = [
        {
            'dictionary_id': 'Word',
            'max_dictionary_size': '50000'

            
        }
    ],
    feature_calcers = [
        'BoW:top_tokens_count=10000'
    ]
)
#bestTest = 0.6601382488
#bestIteration = 5353
# bestTest = 0.6618663594
# bestIteration = 3996
```
